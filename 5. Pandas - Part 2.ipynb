```python
# Low -> Upto 8450
# Medium -> 8450 - 11950
# High -> > 11950
```

```python
for i in range(0, len(cars_data["Price"]), 1):
    if cars_data["Price"][i] <= 8450:
        cars_data["Price_Class"][i] = "Low"
    elif cars_data["Price"][i] > 11950:
        cars_data["Price_Class"][i] = "High"
    else:
        cars_data["Price_Class"][i] = "Medium"
```
